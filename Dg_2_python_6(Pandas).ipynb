```python
#Upper fence and lower fence based on log transformed data

q3 = np.log(df['Fare']+0.1).quantile(0.75)
q1 = np.log(df['Fare']+0.1).quantile(0.25)

upper_fence_log = q3 + (q3 - q1)*1.5
lower_fence_log = q1 - (q3 - q1)*1.5

print(upper_fence_log)
print(lower_fence_log)
```

```python
price_log = np.log(df['Fare']+0.1)

#number of data samples within upper & lower fence if we are considering log transformed value of fare
print(((price_log < upper_fence_log) & (price_log > lower_fence_log)).sum())
print(((price_log < upper_fence_log) & (price_log > lower_fence_log)).sum()/len(df))
```

```python
df_no_outlier_fare = df[(price_log < upper_fence_log) & (price_log > lower_fence_log)]
df_no_outlier_fare.head()
```

**Handling Missing values**

```python
df.isna().sum()
```

```python
df.isna().sum()/len(df)
```

```python
df1 = df.copy()
```

```python
#Drop 'Cabin' column as majority of values are missing in it
#drop() -> to drop or remove a column or a row

df1 = df1.drop('Cabin' , axis = 'columns')
df1.head(5)
```

```python
df1.isna().sum()
```

```python
print(df1['Age'].mean())
print(df1['Age'].median())
```

```python
#In age column approx 19% vlaues are missing, which means majority of data is having age values known
#fill the missing age values with median value of known age
#fillna() - to fill missing values in a data column

df1['Age'] = df1['Age'].fillna(28)
df1.isna().sum()
```

```python
df1.shape
```

```python
#In embarked column we have only two values are missing, which means it is less then 0.5% of total data count
#We can remove those samples with missing values of embarked
#dropna() - to drop a data row with missing values

df1 = df1.dropna(subset = ['Embarked'])
df1.isna().sum()
```

```python
df1.shape
```

* If certain vlaues are missing in a data frame ->
  * we can drop the column - in case majority of values are missing - like we did for cabin column
  * we can fill the gaps with statistically calculated values -> in case majority of values are known , like we did for age column
  * we can drop the data rows - in case a small fraction i.e. lesser the 1% of values ae missing in a data column
